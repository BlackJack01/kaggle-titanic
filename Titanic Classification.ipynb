```python
import pandas as pd
import statsmodels as sm
import numpy as np

import sklearn
from sklearn import cross_validation
from sklearn import tree
from sklearn import svm
from sklearn import ensemble
from sklearn import neighbors
from sklearn import linear_model
from sklearn import metrics
from sklearn import preprocessing
```

```python
df = pd.read_csv("data/train.csv")
# drop the useless columns that we know are not going to be good for prediction
df.drop(["Name","Ticket","Cabin"], axis=1, inplace=True)
```

```python
df = df[np.isfinite(df['Age'])]
df = df[np.isfinite(df['Fare'])]
```

```python
label_encoder = preprocessing.LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])
df['Embarked'] = label_encoder.fit_transform(df['Embarked'])
print df.head()
```

```python
y = df['Survived'].as_matrix().astype(np.int);
```

```python
df.drop(['Survived'], axis=1, inplace=True)
```

```python
print('There are {} instances for survived class and {} instances for died classes.'.format(y.sum(), y.shape[0] - y.sum()))
```

```python
X = df.as_matrix().astype(np.float)
```

```python
# scale your features
scaler = preprocessing.StandardScaler()
X = scaler.fit_transform(X)
```

```python
X
```

```python
def build_cross_validation(X, y, clf_class, shuffle=True, n_folds=3, **kwargs):
    """ run cross validation """
    kfold = cross_validation.StratifiedKFold(y, n_folds=n_folds, shuffle=shuffle)
    y_pred = y.copy()
    for train, test in kfold:
        X_train, X_test = X[train], X[test]
        y_train = y[train]
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_pred[test] = clf.predict(X_test)
    return y_pred
```

```python
print('Passive Aggressive Classifier: {:.2f}'.format(metrics.accuracy_score(y, build_cross_validation(X, y, linear_model.PassiveAggressiveClassifier))))
print('Gradient Boosting Classifier:  {:.2f}'.format(metrics.accuracy_score(y, build_cross_validation(X, y, ensemble.GradientBoostingClassifier))))
print('Support vector machine(SVM):   {:.2f}'.format(metrics.accuracy_score(y, build_cross_validation(X, y, svm.SVC))))
print('Random Forest Classifier:      {:.2f}'.format(metrics.accuracy_score(y, build_cross_validation(X, y, ensemble.RandomForestClassifier))))
print('K Nearest Neighbor Classifier: {:.2f}'.format(metrics.accuracy_score(y, build_cross_validation(X, y, neighbors.KNeighborsClassifier))))
print('Logistic Regression:           {:.2f}'.format(metrics.accuracy_score(y, build_cross_validation(X, y, linear_model.LogisticRegression))))
```

```python
print('Dump Classifier: {:.2f}'.format(metrics.accuracy_score(y, [0 for ii in y.tolist()])))
```

```python
pass_agg_conf_matrix = metrics.confusion_matrix(y, build_cross_validation(X, y, linear_model.PassiveAggressiveClassifier))
grad_ens_conf_matrix = metrics.confusion_matrix(y, build_cross_validation(X, y, ensemble.GradientBoostingClassifier))
decision_conf_matrix = metrics.confusion_matrix(y, build_cross_validation(X, y, tree.DecisionTreeClassifier))
ridge_clf_conf_matrix = metrics.confusion_matrix(y, build_cross_validation(X, y, linear_model.RidgeClassifier))
svm_svc_conf_matrix = metrics.confusion_matrix(y, build_cross_validation(X, y, svm.SVC))
random_forest_conf_matrix = metrics.confusion_matrix(y, build_cross_validation(X, y, ensemble.RandomForestClassifier))
k_neighbors_conf_matrix = metrics.confusion_matrix(y, build_cross_validation(X, y, neighbors.KNeighborsClassifier))
logistic_reg_conf_matrix = metrics.confusion_matrix(y, build_cross_validation(X, y, linear_model.LogisticRegression))
dumb_conf_matrix = metrics.confusion_matrix(y, [0 for ii in y.tolist()]); # ignore the warning as they are all 0

conf_matrix = {
                1: {
                    'matrix': pass_agg_conf_matrix,
                    'title': 'Passive Aggressive',
                   },
                2: {
                    'matrix': grad_ens_conf_matrix,
                    'title': 'Gradient Boosting',
                   },
                3: {
                    'matrix': decision_conf_matrix,
                    'title': 'Decision Tree',
                   },
                4: {
                    'matrix': ridge_clf_conf_matrix,
                    'title': 'Ridge',
                   },
                5: {
                    'matrix': svm_svc_conf_matrix,
                    'title': 'Support Vector Machine',
                   },
                6: {
                    'matrix': random_forest_conf_matrix,
                    'title': 'Random Forest',
                   },
                7: {
                    'matrix': k_neighbors_conf_matrix,
                    'title': 'K Nearest Neighbors',
                   },
                8: {
                    'matrix': logistic_reg_conf_matrix,
                    'title': 'Logistic Regression',
                   },
                9: {
                    'matrix': dumb_conf_matrix,
                    'title': 'Dumb',
                   },
}
```
